```python
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
from sqlalchemy import Column, Integer, String
```

```python
#Creating table Sources
class Sources(Base):
    __tablename__ = 'sources'
    
    JName = Column(String(45), primary_key=True)
    Type = Column(String(45))
    RA = Column(String(45))
    DEC = Column(String(45))

    def __repr__(self):
        return "<members(JName='%s',Type='%s',RA='%s',DEC='%s' )>" % (self.JName,self.Type, self.RA,self.DEC)
```

```python
Sources.__table__
```

```python
Base.metadata.create_all(engine)
```

```python
plsrinfo = [Sources(JName=JName, RA=RAJ, DEC=DECJ) for JName, RAJ,DECJ in psrinfo]
```

```python
from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
session = Session()
```

```python
session.add_all(plsrinfo)
```

```python
session.rollback()
```

```python
p = session.query(Sources).filter_by(JName='J0002+6216').first()
```

```python
p is plsrinfo
```

```python
session.commit()
```

```python
for Sources in session.query(Sources).order_by(Sources.JName):
    print( Sources.JName,Sources.Type, Sources.RA,Sources.DEC)
```
